```python
import os
import random
from tqdm.auto import tqdm

dataset_path = "/content/desktop-tutorial"
new_dataset_path = "/content/desktop-tutorial-10000"

TOTAL_IMAGES = 10000

# Create new dataset folder
os.makedirs(new_dataset_path, exist_ok=True)

# ==========================
# Get all patient folders
# ==========================
patient_folders = sorted(
    [
        folder
        for folder in os.listdir(dataset_path)
        if os.path.isdir(os.path.join(dataset_path, folder))
    ]
)

print(f"Total Patient Folders : {len(patient_folders)}")

images_per_patient = TOTAL_IMAGES // len(patient_folders)

print(f"Images Per Patient : {images_per_patient}")

data = []

# ==========================
# Select Images
# ==========================
for patient in tqdm(patient_folders, desc="Selecting Images"):

    patient_path = os.path.join(dataset_path, patient)

    patient_images = []

    # Collect images from class 0 and class 1
    for label in ["0", "1"]:

        label_path = os.path.join(patient_path, label)

        if not os.path.exists(label_path):
            continue

        for file in os.listdir(label_path):

            if file.endswith(".png"):

                patient_images.append(
                    (
                        os.path.join(label_path, file),
                        int(label),
                        patient,
                        file
                    )
                )

    # Shuffle images
    random.shuffle(patient_images)

    # Select images
    selected_images = patient_images[:images_per_patient]

    # Copy images
    for image_path, label, patient, file in selected_images:

        destination_folder = os.path.join(
            new_dataset_path,
            patient,
            str(label)
        )

        os.makedirs(destination_folder, exist_ok=True)

        shutil.copy2(
            image_path,
            os.path.join(destination_folder, file)
        )

        data.append([

            os.path.join(destination_folder, file),

            label
        ])

# ==========================
# Create DataFrame
# ==========================
df_small = pd.DataFrame(
    data,
    columns=["image_path", "label"]
)

print("\nDataset Created Successfully")
print(df_small.shape)

print(df_small.head())

# ==========================
# Save CSV
# ==========================
csv_path = os.path.join(new_dataset_path, "selected_10000.csv")

df_small.to_csv(csv_path, index=False)

print(f"\nCSV Saved : {csv_path}")
print(f"New Dataset Location : {new_dataset_path}")
```

```python
import os
import gc
import cv2
import numpy as np
from tqdm.auto import tqdm

IMG_SIZE = 224
CHUNK_SIZE = 1500

save_dir = "/content/numpy_chunks"
os.makedirs(save_dir, exist_ok=True)

images = []
labels = []

chunk = 0

for _, row in tqdm(df.iterrows(), total=len(df), desc="Processing Images"):

    image_path = row["image_path"]

    img = cv2.imread(image_path)

    if img is None:
        continue

    # Convert BGR → RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Resize
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

    # Normalize
    img = img.astype(np.float32) / 255.0

    images.append(img)
    labels.append(row["label"])

    # Save every CHUNK_SIZE images
    if len(images) == CHUNK_SIZE:

        X = np.array(images, dtype=np.float32)
        y = np.array(labels)

        np.save(os.path.join(save_dir, f"X_chunk_{chunk}.npy"), X)
        np.save(os.path.join(save_dir, f"y_chunk_{chunk}.npy"), y)

        print(f"Saved Chunk {chunk} --> {X.shape}")

        # Clear RAM
        del X
        del y
        images.clear()
        labels.clear()

        gc.collect()

        chunk += 1

# Save remaining images
if len(images) > 0:

    X = np.array(images, dtype=np.float32)
    y = np.array(labels)

    np.save(os.path.join(save_dir, f"X_chunk_{chunk}.npy"), X)
    np.save(os.path.join(save_dir, f"y_chunk_{chunk}.npy"), y)

    print(f"Saved Final Chunk {chunk} --> {X.shape}")

    del X
    del y
    gc.collect()

print("\nAll images processed successfully.")
```

```python
import os
import gc
import numpy as np
from sklearn.model_selection import train_test_split

chunk_folder = "/content/numpy_chunks"

# Create folders
train_folder = "/content/train_chunks"
test_folder = "/content/test_chunks"

os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# Get all chunk numbers
chunk_numbers = sorted(
    int(f.split("_")[-1].split(".")[0])
    for f in os.listdir(chunk_folder)
    if f.startswith("X_chunk_")
)

print(f"Total Chunks: {len(chunk_numbers)}")

for i in chunk_numbers:

    print(f"\nProcessing Chunk {i}")

    # Load chunk
    X = np.load(os.path.join(chunk_folder, f"X_chunk_{i}.npy"))
    y = np.load(os.path.join(chunk_folder, f"y_chunk_{i}.npy"))

    # Split chunk
    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=0.20,
        random_state=42,
        stratify=y
    )

    # Save train split
    np.save(os.path.join(train_folder, f"X_train_{i}.npy"), X_train)
    np.save(os.path.join(train_folder, f"y_train_{i}.npy"), y_train)

    # Save test split
    np.save(os.path.join(test_folder, f"X_test_{i}.npy"), X_test)
    np.save(os.path.join(test_folder, f"y_test_{i}.npy"), y_test)

    print(f"Train : {X_train.shape}")
    print(f"Test  : {X_test.shape}")

    # Free memory
    del X, y, X_train, X_test, y_train, y_test
    gc.collect()

print("\nAll chunks split successfully!")
```

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import (
    GlobalAveragePooling2D,
    BatchNormalization,
    Dense,
    Dropout
)
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2

# ==================================================
# Load Pretrained MobileNetV2
# ==================================================

base_model = MobileNetV2(
    weights="imagenet",
    include_top=False,
    input_shape=(224, 224, 3)
)

# Freeze pretrained layers
base_model.trainable = False

# ==================================================
# Build Custom Classifier
# ==================================================

x = base_model.output

x = GlobalAveragePooling2D()(x)

x = BatchNormalization()(x)

x = Dense(
    128,
    activation="relu",
    kernel_regularizer=l2(0.001)
)(x)

x = Dropout(0.5)(x)

x = Dense(
    64,
    activation="relu",
    kernel_regularizer=l2(0.001)
)(x)

x = Dropout(0.3)(x)

output = Dense(
    1,
    activation="sigmoid"
)(x)

# ==================================================
# Create Model
# ==================================================

model = Model(
    inputs=base_model.input,
    outputs=output
)

# ==================================================
# Compile Model
# ==================================================

model.compile(
    optimizer=tf.keras.optimizers.Adam(
        learning_rate=1e-4
    ),
    loss="binary_crossentropy",
    metrics=[
        "accuracy",
        tf.keras.metrics.Precision(name="precision"),
        tf.keras.metrics.Recall(name="recall"),
        tf.keras.metrics.AUC(name="auc")
    ]
)

# ==================================================
# Display Model Summary
# ==================================================

model.summary()
```

```python
import os
import gc
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.models import load_model
from tensorflow.keras.optimizers import Adam

# ==========================================================
# PATHS
# ==========================================================

MODEL_PATH = "/content/best_model.keras"

TRAIN_FOLDER = "/content/train_chunks"

TEST_FOLDER = "/content/test_chunks"

SAVE_MODEL = "/content/best_finetuned_model.keras"

EPOCHS = 25

BATCH_SIZE = 32


# ==========================================================
# CHECK PATHS
# ==========================================================

print("="*70)
print("Checking Files...")
print("="*70)

if not os.path.exists(MODEL_PATH):

    print("\nFiles inside /content :")
    print(os.listdir("/content"))
    print(f"\nModel not found at {MODEL_PATH}. Saving the current model to this path to proceed with fine-tuning.")
    # Save the existing 'model' object (from previous cells) to resolve FileNotFoundError
    model.save(MODEL_PATH)
    print("Current model saved successfully to best_model.keras.")

if not os.path.exists(TRAIN_FOLDER):

    raise FileNotFoundError(f"{TRAIN_FOLDER} not found.")

if not os.path.exists(TEST_FOLDER):

    raise FileNotFoundError(f"{TEST_FOLDER} not found.")

# ==========================================================
# LOAD MODEL
# ==========================================================

print("\nLoading Best Model...\n")

model = load_model(MODEL_PATH)

print("Model Loaded Successfully")

# ==========================================================
# FIND GAP LAYER
# ==========================================================

gap_index = -1

for i, layer in enumerate(model.layers):

    if "global_average_pooling2d" in layer.name:

        gap_index = i

        break

if gap_index == -1:

    raise ValueError("GlobalAveragePooling2D layer not found.")

# ==========================================================
# UNFREEZE LAST 30 MOBILENET LAYERS
# ==========================================================

mobile_layers = model.layers[1:gap_index]

for layer in mobile_layers:

    layer.trainable = True

freeze_until = len(mobile_layers) - 30

if freeze_until > 0:

    for layer in mobile_layers[:freeze_until]:

        layer.trainable = False

for layer in model.layers[gap_index:]:

    layer.trainable = True

print("\nTrainable Layers\n")

for layer in model.layers:

    if layer.trainable:

        print(layer.name)

# ==========================================================
# COMPILE
# ==========================================================

model.compile(

    optimizer=Adam(learning_rate=1e-5),

    loss="binary_crossentropy",

    metrics=[

        "accuracy",

        tf.keras.metrics.Precision(name="precision"),

        tf.keras.metrics.Recall(name="recall"),

        tf.keras.metrics.AUC(name="auc")

    ]

)

# ==========================================================
# FIND TRAIN CHUNKS
# ==========================================================

chunk_numbers = sorted([

    int(f.split("_")[-1].split(".")[0])

    for f in os.listdir(TRAIN_FOLDER)

    if f.startswith("X_train_") and f.endswith(".npy")

])

if len(chunk_numbers) == 0:

    raise ValueError("No training chunks found.")

print("\nTotal Training Chunks :", len(chunk_numbers))

# ==========================================================
# LOAD COMPLETE VALIDATION DATA
# ==========================================================

print("\nLoading Validation Dataset...\n")

X_val_list = []

y_val_list = []

for chunk in chunk_numbers:

    x_file = os.path.join(TEST_FOLDER, f"X_test_{chunk}.npy")

    y_file = os.path.join(TEST_FOLDER, f"y_test_{chunk}.npy")

    if not os.path.exists(x_file):

        print(f"Skipping Validation Chunk {chunk}")

        continue

    X_val_list.append(np.load(x_file))

    y_val_list.append(np.load(y_file))

if len(X_val_list) == 0:

    raise ValueError("Validation dataset is empty.")

X_val = np.concatenate(X_val_list, axis=0)

y_val = np.concatenate(y_val_list, axis=0)

print("Validation Shape :", X_val.shape)

# ==========================================================
# HISTORY
# ==========================================================

fine_train_acc = []

fine_val_acc = []

fine_train_loss = []

fine_val_loss = []

best_val_loss = float("inf")

patience = 3

wait = 0

min_delta = 0.001
# ==========================================================
# START FINE-TUNING
# ==========================================================

for epoch in range(EPOCHS):

    print("\n" + "=" * 70)
    print(f"Fine-Tuning Epoch {epoch + 1}/{EPOCHS}")
    print("=" * 70)

    epoch_train_acc = []
    epoch_train_loss = []

    # ======================================================
    # TRAIN ON EACH TRAINING CHUNK
    # ======================================================

    for chunk in chunk_numbers:

        print(f"\nTraining Chunk {chunk}")

        x_file = os.path.join(TRAIN_FOLDER, f"X_train_{chunk}.npy")
        y_file = os.path.join(TRAIN_FOLDER, f"y_train_{chunk}.npy")

        if not os.path.exists(x_file):

            print(f"Skipping Chunk {chunk}")

            continue

        X_train = np.load(x_file)
        y_train = np.load(y_file)

        history = model.fit(

            X_train,
            y_train,

            epochs=1,

            batch_size=BATCH_SIZE,

            verbose=1,

            shuffle=True

        )

        epoch_train_acc.append(history.history["accuracy"][-1])
        epoch_train_loss.append(history.history["loss"][-1])

        del X_train
        del y_train

        gc.collect()

    # ======================================================
    # TRAINING METRICS
    # ======================================================

    avg_train_acc = np.mean(epoch_train_acc)
    avg_train_loss = np.mean(epoch_train_loss)

    # ======================================================
    # VALIDATION
    # ======================================================

    print("\nEvaluating Validation Dataset...")

    results = model.evaluate(

        X_val,
        y_val,

        batch_size=BATCH_SIZE,

        verbose=1

    )

    val_loss = results[0]
    val_accuracy = results[1]
    val_precision = results[2]
    val_recall = results[3]
    val_auc = results[4]

    fine_train_acc.append(avg_train_acc)
    fine_train_loss.append(avg_train_loss)

    fine_val_acc.append(val_accuracy)
    fine_val_loss.append(val_loss)

    # ======================================================
    # DISPLAY RESULTS
    # ======================================================

    print("\n" + "=" * 60)
    print("Epoch Summary")
    print("=" * 60)

    print(f"Training Accuracy   : {avg_train_acc:.4f}")
    print(f"Validation Accuracy : {val_accuracy:.4f}")

    print(f"Training Loss       : {avg_train_loss:.4f}")
    print(f"Validation Loss     : {val_loss:.4f}")

    print(f"Precision           : {val_precision:.4f}")
    print(f"Recall              : {val_recall:.4f}")
    print(f"AUC                 : {val_auc:.4f}")

    # ======================================================
    # EARLY STOPPING
    # ======================================================

    if val_loss < (best_val_loss - min_delta):

        best_val_loss = val_loss

        wait = 0

        model.save(SAVE_MODEL)

        print("\nValidation Loss Improved")
        print("Best Fine-Tuned Model Saved")

    else:

        wait += 1

        print(f"\nNo Improvement ({wait}/{patience})")

        if wait >= patience:

            print("\nEarly Stopping Triggered")

            break

# ==========================================================
# LOAD BEST MODEL
# ==========================================================

if os.path.exists(SAVE_MODEL):

    model = load_model(SAVE_MODEL)

    print("\nBest Fine-Tuned Model Loaded")

# ==========================================================
# PLOT ACCURACY
# ==========================================================

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)

plt.plot(

    range(1,len(fine_train_acc)+1),

    fine_train_acc,

    marker="o",

    linewidth=2,

    label="Training Accuracy"

)

plt.plot(

    range(1,len(fine_val_acc)+1),

    fine_val_acc,

    marker="o",

    linewidth=2,

    label="Validation Accuracy"

)

plt.title("Fine-Tuning Accuracy")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.grid(True)

plt.legend()

# ==========================================================
# PLOT LOSS
# ==========================================================

plt.subplot(1,2,2)

plt.plot(

    range(1,len(fine_train_loss)+1),

    fine_train_loss,

    marker="o",

    linewidth=2,

    label="Training Loss"

)

plt.plot(

    range(1,len(fine_val_loss)+1),

    fine_val_loss,

    marker="o",

    linewidth=2,

    label="Validation Loss"

)

plt.title("Fine-Tuning Loss")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.grid(True)

plt.legend()

plt.tight_layout()

plt.show()

# ==========================================================
# FINAL RESULTS
# ==========================================================

print("\n" + "=" * 60)
print("Fine-Tuning Completed Successfully")
print("=" * 60)

print(f"Best Validation Loss : {best_val_loss:.4f}")

if len(fine_val_acc) > 0:

    print(f"Final Validation Accuracy : {fine_val_acc[-1]:.4f}")
    print(f"Final Training Accuracy   : {fine_train_acc[-1]:.4f}")

else:

    print("Training did not complete.")
```

```python
import os
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

test_folder = "/content/test_chunks"

# Get all chunk numbers
chunk_numbers = sorted(
    int(f.split("_")[-1].split(".")[0])
    for f in os.listdir(test_folder)
    if f.startswith("X_test_")
)

y_true_all = []
y_pred_all = []

for chunk in chunk_numbers:

    print(f"Evaluating Chunk {chunk}")

    # Load chunk
    X_test = np.load(os.path.join(test_folder, f"X_test_{chunk}.npy"))
    y_test = np.load(os.path.join(test_folder, f"y_test_{chunk}.npy"))

    # Predict
    y_pred = model.predict(X_test, verbose=0)

    # Convert probabilities to 0/1
    y_pred = (y_pred > 0.5).astype(int).flatten()

    # Store results
    y_true_all.extend(y_test)
    y_pred_all.extend(y_pred)

# Convert to numpy arrays
y_true_all = np.array(y_true_all)
y_pred_all = np.array(y_pred_all)

# Confusion Matrix
cm = confusion_matrix(y_true_all, y_pred_all)

plt.figure(figsize=(6,5))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Benign", "Malignant"],
    yticklabels=["Benign", "Malignant"]
)

plt.title("Confusion Matrix (All Test Chunks)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
print("\nClassification Report\n")
print(classification_report(
    y_true_all,
    y_pred_all,
    target_names=["Benign", "Malignant"]
))
```
